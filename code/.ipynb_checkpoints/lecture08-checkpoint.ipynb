Exercise: Write a Python function that takes a matrix, $A$, and a
vector, $b$, and computes the solution to the linear system
$A^TAx = b$.  If $A$ has $n$ rows and $m$ columns, $A^TA$ will
be $m$ by $m$.  You can assume that $b$ has $m$ elements.

```python
n = 10
m = 4
A = numpy.random.normal(size=(n, m))
M = A.transpose().dot(A)
M
```

Make $b$

```python
b = numpy.arange(4)
b
```

Solve

```python
x = numpy.linalg.solve(A.transpose().dot(A), b)
x
```

Check the solution.

```python
print(M.dot(x) - b)
```

Exercise: Spectrum provides an '__add__' method.
If I take the spectrum of the violin sample and the spectrum
of the saxophone sample, add them using this function, and then
compute the corresponding wave, what would it sound like?

```python
wave1 = thinkdsp.read_wave('92002__jcveliz__violin-origional.wav')
wave2 = thinkdsp.read_wave('100475__iluppai__saxophone-weep.wav')
```

Play the violin sample:

```python
wave1.make_audio()
```

And the saxophone sample:

```python
wave2.make_audio()
```

Trim them to the same length:

```python
n = min(len(wave1), len(wave2))
wave1.ys = wave1.ys[:n]
wave2.ys = wave2.ys[:n]
```

Check that the framerates are equal:

```python
wave1.framerate, wave2.framerate
```

Play them together:

```python
n = max(len(wave2), len(wave1))
ys = wave2.ys.copy()
#ys.resize(n)
ys += wave1.ys
ys
```

```python
wave = wave1 + wave2
wave.make_audio()
```

Add up their spectrums:

```python
spectrum = wave1.make_spectrum() + wave2.make_spectrum()
```

Convert back to a wave and play it.

```python
spectrum.make_wave().make_audio()
```

It seems like adding things up in the frequency domain is the same as adding them up in the time domain.

That's true because the DFT is linear, so the DFT(f+g) = DFT(f) + DFT(g).

What happens if we add two waves that are out of phase by 180 degrees?

```python
wave1 = thinkdsp.TriangleSignal(freq=400).make_wave(duration=1)
wave1.make_audio()
```

```python
wave2 = thinkdsp.TriangleSignal(freq=400, offset=math.pi).make_wave(duration=1)
wave2.make_audio()
```

Here's the sum in the time domain:

```python
wave = thinkdsp.Wave(wave1.ys + wave2.ys, wave1.framerate)
wave.make_audio()
```

And here's the sum in the frequency domain:

```python
spectrum = wave1.make_spectrum() + wave2.make_spectrum()
spectrum.make_wave().make_audio()
```

Either way, the waves cancel each other out.
