Problem 6

Write a class called SawtoothSignal that extends Signal and provides evaluate to evaluate a sawtooth signal.  Note: there is an implementation of SawtoothSignal in thinkdsp.py; you should avoid looking at it until you have written your own.

```python
import math

class sawtoothSignal(thinkdsp.Sinusoid):
    def evaluate(self, ts):
        cycles = self.freq * ts + self.offset / math.pi*2
        frac, _ = numpy.modf(cycles)
        ys = thinkdsp.normalize(thinkdsp.unbias(frac), self.amp)
        return ys

sig = sawtoothSignal(500)
duration = sig.period * 5
seg = sig.make_wave(duration, framerate=10000)
seg.plot()        
```

Problem 10

The triangle and square waves have odd harmonics only; the sawtooth wave has both even and odd harmonics. The harmonics of the square and sawtooth waves drop off in proportion to 1/f; the harmonics of the triangle wave drop off like 1/f2. Can you find a waveform that has even and odd harmonics that drop off like 1/f2?

```python
cos_sig = 0
for f in range (100,5000,100):
    cos_sig += thinkdsp.CosSignal(freq=f, amp=-1/(f*f), offset=0)
    
mix_wave = cos_sig.make_wave(duration=0.06, start=0, framerate=10000)
mix_wave.plot()
```

```python
mix_spec = mix_wave.make_spectrum()
mix_spec.plot()
```
