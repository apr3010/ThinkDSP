Problem 2

Write a class called SawtoothChirp that extends Chirp and overrides evaluate to generate a sawtooth waveform with frequency that increases (or decreases) linearly. Plot the spectrogram of this signal. The effect of aliasing should be visible, and if you listen carefully, you can hear it.

```python
import math
import numpy
PI2 = math.pi*2
class SawtoothChirp(thinkdsp.Chirp):
 
    def _evaluate(self, ts, freqs):
        dts = numpy.diff(ts)
        dps = PI2 * freqs * dts
        phases = numpy.cumsum(dps)
        phases = numpy.insert(phases, 0, 0)
        cycles = phases / PI2
        frac, _ = numpy.modf(cycles)
        ys = thinkdsp.normalize(thinkdsp.unbias(frac), self.amp)
        return ys

sig = SawtoothChirp(start=220,end=880)
seg = sig.make_wave(duration=0.03)
seg.plot()
```

```python
segment = sig.make_wave(duration=1)
spectro = segment.make_spectrogram(512)
spectro.plot()
```

Problem 3

In musical terminology, a “glissando” is a note that slides from one pitch to another, so it is similar to a chirp. A trombone player can play a glissando by extending the trombone slide while blowing continuously. As the slide extends, the total length of the tube gets longer, and the resulting pitch is inversely proportional to length.

Assuming that the player moves the slide at a constant speed, how does frequency vary with time? Is a trombone glissando more like a linear or exponential chirp?

Write a function that simulates a trombone glissando from C3 up to F3 and back down to C3. C3 is 262 Hz; F3 is 349 Hz.

```python
class Glissando(thinkdsp.Chirp):
    def evaluate(self, ts):
        start, end = 1.0/self.start, 1.0/self.end
        pd = numpy.linspace(start, end, len(ts)-1)
        freqs = 1/pd
        return self._evaluate(ts,freqs)
```

A trombone glissando is more like a linear chirp, because the perceived pitch varies linearly and the frequency varies exponentially with time.

```python
sig = Glissando(262, 349)
trom = sig.make_wave(duration=1)
trom.make_audio()
```

```python
sig2 = Glissando(349, 262)
trom2 = sig2.make_wave(duration=1)
trom2.make_audio()
```

```python
trombone = trom | trom2
trombone.make_audio()
```

```python
spectrogram = trombone.make_spectrogram(512)
spectrogram.plot(high=30)
```

In a white noise signal, the mixture of frequencies changes over time. In the long run, we expect the power at all frequencies to be equal, but in any sample, the power at each frequency is random.
To estimate the long-term average power at each frequency, we can break a long signal into segments, compute the power spectral density for each segment, and then compute the average across the segments. 

```python
import matplotlib.pyplot as plt
import numpy
step = 1
start = 0
duration = 3
a = []
pwr = 0
for K in range(start, duration, step):
    new = bowl.segment(K, step)
    w = new.make_spectrum()
    pwr += w.power
    avg = pwr/duration
plt.plot(avg,'-')
thinkplot.config(xlim=[0, 200], ylim=[0, 1.4e20], legend=False)
```

Problem 7

At http://www.coindesk.com you can download the daily price of a BitCoin as a CSV file.  Read this file and compute the power spectrum of BitCoin prices as a function of time. Does it resemble white, pink, or Brownian noise?

```python
import csv
import numpy
import matplotlib.pyplot as plt
t = 0
with open('coindesk-bpi-USD-close.csv') as f:
    reader = csv.reader(f)
    for row in reader:
        t += 1
        plt.plot(t,row[1],'x')
thinkplot.show(xlabel='frequency',ylabel='power', xscale='log',yscale='log')
```

The beta factor is less than 0, so it is neither white, red nor pink noise. The plot actually resembles blue noise, where the power density increases +3 dB/octave.
