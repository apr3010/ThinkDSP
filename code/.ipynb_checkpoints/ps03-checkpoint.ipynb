2) Compute the autocorrelation of BitCoin prices. Does the autocorrelation function decay exponentially? Is there evidence of periodic behavior?

Looking at the graph, it shows that as the lag increases the correlation decreases. The autocorrelation function is somewhat exponential, because the rate at which the correlation decreases is greater at the beginning than at the end.

There is no point at which the correlation increases back to 1, meaning that the wave has not aligned with itself ever again. Since the wave has never replicated itself again, there is no obvious periodic behaviour.

```python
def corrcoef(xs, ys):
    return numpy.corrcoef(xs, ys, ddof=0)[0,1]

def serial_corr(wave, lag=1):
    n = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:n-lag]
    corr = numpy.corrcoef(y1, y2)[0,1]
    return corr

def autocorr(wave):
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave,lag) for lag in lags]
    return lags, corrs

lags, corrs = autocorr(wave1)
thinkplot.plot(lags,corrs)
thinkplot.show(xlabel='Lag', ylabel='Correlation')
```

When comparing the running time of analyze1 with analyze2, analyze2 is much more efficient. The function analyze2 uses the dot product, whereas analyze1 uses linalg.solve to compute the results. Evidently, using the dot product has a lower complexity and is therefore faster to run. Scipy.fftpack.dct runs much faster than both functions, because the algorithm used to compute the amplitudes is much faster and must have a very low complexity.

```python
import scipy

PI2 = 2 * math.pi
signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=1.0)

ns = 2 ** numpy.arange(4, 12)
print(ns)
results = []
```

```python
results = []
for N1 in ns:
    ts = (0.5 + numpy.arange(N1)) / N1
    freqs = (0.5 + numpy.arange(N1)) / 2
    ys = noise.ys[:N1]
    result = %timeit -o dct.analyze1(ys, freqs, ts)
    results.append(result)
```

5) One of the major applications of the DCT is compression for both sound and images. In its simplest form, DCT-based compression works like this:
- Break a long signal into segments.
- Compute the DCT of each segment.
- Identify frequency components with amplitudes so low they are inaudible, and remove them. Store only the frequencies and amplitudes that remain.
- To play back the signal, load the frequencies and amplitudes for each segment and apply the inverse DCT.
Implement a version of this algorithm and apply it to a recording of music or speech. How many components can you eliminate before the difference is perceptible?

I kept increasing the limit and I started noticing a difference in the audio file when I eliminated values under 1000000. The difference I noticed was a higher pitch since low amplitudes of non-significant frequencies were eliminated. Apparently, this indicates that the higher frequencies were more dominant than the lower ones, resulting in a higher perceived pitch.

```python
sig = thinkdsp.read_wave('crickets.wav')
sig_seg = sig.segment(0,0)

sig_sg = sig.segment(0,10)
sig_sg.make_audio() # Original segment to compare to final version
sig_sg.plot()
thinkplot.show(xlim=[0.4,0.45])
```

```python
final = sig_seg
step = 1
start = 0
duration = 10

def check(element):
    if abs(element) > 1000000:
        return element
    return 0

for K in range(start, duration, step):
    new = sig.segment(K, step) 
    new = new.make_spectrum() # uses DCT
    new.hs = map(check, new.hs) # filters unwanted hs
    final = final.__or__(new.make_wave()) # adds waves together
final.make_audio()
```

```python
final.plot()
thinkplot.show(xlim=[0.4,0.45])
```
