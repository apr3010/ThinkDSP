2) Read the Wikipedia page on DFT.  Write an implementation of DFT based on this summation and compare your results to my implementation and/or numpy.fft.fft.

N = the number of samples (same as the length of the ys array of the wave)

k = frequency = samples per second

n = sample you are at

The DFT converts the wave into a spectrum, determining how much of each frequency exists in the wave. For every frequency (that apparently exists within a range of 0 to 4 since it is "normalized"), all the samples are run and the sum of the equation from the Wikipedia page is calculated. The final sum represents the amplitude for a particular frequency.

```python
signal = thinkdsp.ComplexSinusoid(freq=100, amp=0.6, offset=1)
wave = signal.make_wave(duration=1, framerate=4)
print(wave.ys)
```

```python
amps = []

def mydft(ys):
    N = len(ys)
    for k in range (4):
        a = 0
        for n in range (N):
            a += ys[n]*numpy.exp(-PI2*k*n*i/N)
        amps.append(a)
    return amps
mydft(wave.ys)
```

```python
def dft(ys):
    N = len(ys)
    ts = numpy.arange(N) / N
    freqs = numpy.arange(N)
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    amps = M.conj().transpose().dot(ys)
    return amps
dft(wave.ys)
```

3) In Chapter 7, I showed how we can express the DFT and inverse DFT as matrix multiplications. These operations are relatively fast, taking time proportional to N2, where N is the length of the wave array. That would be fast enough for many applications, but it turns out that there is a faster algorithm, the Fast Fourier Transform (FFT), which takes time proportional to N logN.

Read about the FFT at http://en.wikipedia.org/wiki/Cooley-Tukey_FFT_algorithm, and write an implementation. Hint: I suggest you write a simple version as a recursive function; don’t worry about “data reordering, bit reversal, and in-place algorithms”.

Even though I was able to understand how the algorithm worked, I could not really figure out how to combine the even and odd sections together, so I looked at the following website for coding help: http://jeremykun.com/2012/07/18/the-fast-fourier-transform/. Basically, what the code does is goes to the base case, where N, the number of samples, is equal to 1 and it builds the even and odd sections from there. Dividing the wave into two sections makes the fft a lot faster to compute. The even section does the fft on all the even components of the array inputed and the odd section does the same for the odd components. For each section, a range of frequencies are run, using the equations on the wikipedia site, to combine the even and odd equations calculated. An empty array of length N is generated so that the combined values can be overwritten on it. One equation is in the even elements of the array, but is written in all the elements of the array, so the second equation overwrites all the odd elements of the array with the second equation.

```python
import math
def omega(a, b):
   return numpy.exp((-PI2 * 1j * b) / a)

def fft(sig):
    N = len(sig)
    if N == 1:
        return sig
    else:
        evens = fft([sig[i] for i in xrange(0, N, 2)])
        odds = fft([sig[i] for i in xrange(1, N, 2)])

        combined = [0] * N

    for k in xrange(int(N/2)):
        combined[k] = evens[k] + omega(N, k) * odds[k]
        combined[k + int(N/2)] = evens[k] - omega(N, k) * odds[k]

    return combined
fft(wave.ys)
```

4) Using the timing code from PS03, compare the performance of my implementation of autocorr from Chapter 5, my fft_autocorr from Chapter 8, and numpy.correlate.  Can you tell whether NumPy is using the FFT algorithm?

```python
import scipy

PI2 = 2 * math.pi
signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=1.0)

ns = 2 ** numpy.arange(4, 12)
print(ns)
results = []
```

```python
def corrcoef(xs, ys):
    return numpy.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr

def autocorr(wave):
    """Computes and plots the autocorrelation function.

    wave: Wave
    """
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs
```

```python
results = []
for N1 in ns:
    ys = noise.ys[:N1]
    wv = thinkdsp.Wave(ys, framerate=11025)
    result = %timeit -o autocorr(wv)
    results.append(result)

best1 = [result.best for result in results]
```

```python
def zero_pad(array, n):
    res = numpy.zeros(n)
    res[:len(array)] = array
    return res

def fft_convolve(signal, window):
    fft_signal = numpy.fft.fft(signal)
    fft_window = numpy.fft.fft(window)
    return numpy.fft.ifft(fft_signal * fft_window)

def fft_autocorr(signal):
    N = len(signal)
    window = signal[::-1]
    signal = zero_pad(signal, 2*N)
    window = zero_pad(window, 2*N)

    corrs = fft_convolve(signal, window)
    corrs = corrs[N//2: 3*N//2]
    return corrs
```

```python
results = []
for N2 in ns:
    ys = noise.ys[:N2] 
    result = %timeit -o fft_autocorr(ys)
    results.append(result)

best2 = [result.best for result in results]
```

```python
results = []
for N3 in ns:
    ys = noise.ys[:N3] 
    result = %timeit -o numpy.correlate(ys, ys)
    results.append(result)
    
best3 = [result.best for result in results]
```

```python
thinkplot.plot(ns, best1, label='Chap5_autocorr')
thinkplot.plot(ns, best2, label='fft_autocorr')
thinkplot.plot(ns, best3, label='numpy.correlate')
thinkplot.config(xscale='log', yscale='log', xlabel='Loops', ylabel='Run Time')
```

```python
x1 = numpy.log(ns)
y1 = numpy.log(best1)
t1 = scipy.stats.linregress(x1,y1)
print(t1[0])

y2 = numpy.log(best2)
t2 = scipy.stats.linregress(x1,y2)
print(t2[0])

y3 = numpy.log(best3)
t3 = scipy.stats.linregress(x1,y3)
print(t3[0])
```

Looking at these results, we can tell that Allen's code takes the longest time to run and is the least efficient out of the three. Numpy works the fastest, faster than the fft and Allen's code. Also, for Allen's code, the number of loops has a proportional relationship to the run time, but for numpy.correlate, increasing the loops doesn't affect the run time as much. Since numpy.correlate is much faster than the fft algorithm, it cannot be using the algorithm, because otherwise, numpy.correlate would be slower.
