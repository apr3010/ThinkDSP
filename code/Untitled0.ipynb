```python
from __future__ import print_function, division

import thinkdsp
import thinkplot
import dct
import thinkstats2
import math
import pandas
import numpy
import matplotlib.pyplot as plt

%precision 3
%matplotlib inline
i = complex(0,1)
```

```python
PI2 = 2*math.pi
signal = thinkdsp.ComplexSinusoid()
wave = signal.make_wave()
spectrum = wave.make_spectrum()
signal.plot()
```

```python
factors = PI2 * i *spectrum.fs
spectrum.hs *= factors
diff = spectrum.make_wave()
diff.plot()
thinkplot.config(xlim=[0, 0.1])
```
